```python
class CandidateFunctions:
    # Linear function
    def linear(self, x, a, b):
        return a * x + b
    
    # Quadratic function
    def quadratic(self, x, a, b, c):
        return a * x**2 + b * x + c
    
    # Cubic function
    def cubic(self, x, a, b, c, d):
        return a * x**3 + b * x**2 + c * x + d
    
    # Quartic function
    def quartic(self, x, a, b, c, d, e):
        return a * x**4 + b * x**3 + c * x**2 + d * x + e
    
    # Exponential function
    def exponential(self, x, a, b, c):
        return a * np.exp(b * x) + c
    
    # Logarithmic function
    def logarithmic(self, x, a, b, c):
        return a * np.log(b * x) + c
    
    # Power Law function
    def power_law(self, x, a, b):
        return a * x**b
    
    # Sigmoid/Logistic function
    def sigmoid(self, x, L, k, x0):
        return L / (1 + np.exp(-k * (x - x0)))
    
    # Hyperbolic Tangent function
    def hyperbolic_tangent(self, x, a, b, c):
        return a * np.tanh(b * x) + c
    
    # Sinusoidal function
    def sinusoidal(self, x, a, b, c, d):
        return a * np.sin(b * x + c) + d
    
    # Gaussian function
    def gaussian(self, x, a, b, c):
        return a * np.exp(-((x - b)**2) / (2 * c**2))
    
    # Rational function
    def rational(self, x, a, b, c, d):
        return (a * x + b) / (c * x + d)
    
    # Inverse function
    def inverse(self, x, a, b):
        return a / x + b
    
    # Square Root function
    def square_root(self, x, a, b):
        return a * np.sqrt(x) + b
    
    # Absolute Value function
    def absolute_value(self, x, a, b):
        return a * np.abs(x) + b
    
    # Beta function (defined for x in [0, 1])
    def beta_function(self, x, a, b):
        return x**a * (1 - x)**b
    
    # Weibull function
    def weibull(self, x, k, lamb):
        return (k / lamb) * (x / lamb)**(k - 1) * np.exp(-(x / lamb)**k)
    
    # Gompertz function
    def gompertz(self, x, a, b, c):
        return a * np.exp(-b * np.exp(-c * x))
    
    # Polynomial of Degree n (for example, n=5)
    def polynomial_n(self, x, *coefficients):
        return sum(c * x**i for i, c in enumerate(coefficients))
    
    # Dictionary of functions with initial parameter guesses
    def get_functions(self):
        return {
            'linear': (self.linear, [1, 0]),
            'quadratic': (self.quadratic, [1, 0, 0]),
            'cubic': (self.cubic, [1, 0, 0, 0]),
            'quartic': (self.quartic, [1, 0, 0, 0, 0]),
            'exponential': (self.exponential, [1, 0.1, 1]),
            'logarithmic': (self.logarithmic, [1, 1, 1]),
            'power_law': (self.power_law, [1, 0.1]),
            'sigmoid': (self.sigmoid, [1, 1, 0]),
            'hyperbolic_tangent': (self.hyperbolic_tangent, [1, 1, 0]),
            'sinusoidal': (self.sinusoidal, [1, 1, 0, 0]),
            'gaussian': (self.gaussian, [1, 0, 1]),
            'rational': (self.rational, [1, 0, 1, 0]),
            'inverse': (self.inverse, [1, 0]),
            'square_root': (self.square_root, [1, 0]),
            'absolute_value': (self.absolute_value, [1, 0]),
            'beta_function': (self.beta_function, [1, 1]),
            'weibull': (self.weibull, [1, 1]),
            'gompertz': (self.gompertz, [1, 1, 1]),
            'polynomial_5': (self.polynomial_n, [1, 0, 0, 0, 0, 0]),  # Quintic
            'polynomial_6': (self.polynomial_n, [1, 0, 0, 0, 0, 0, 0]),  # Sextic
        }
        
```

```python
# Get functions
instance = CandidateFunctions()
funcs = instance.get_functions()
```

```python
plt.scatter(x_values, y_values, label="Real Data")
fit_list = []
for name, (func, initial_guess) in funcs.items():
    try:
        # Fit function to data
        params, covariance = curve_fit(func, x_values, y_values, p0=initial_guess)
        fitted_curve = func(x_values, *params)
        # Calculate R-squared
        residuals = y_values - fitted_curve
        ss_res = np.sum(residuals**2)
        ss_tot = np.sum((y_values - np.mean(y_values))**2)
        r_squared = 1 - (ss_res / ss_tot)
        if r_squared > 0.99:
            plt.plot(x_values, fitted_curve, label=name)
            fit_list.append((name, r_squared, params, fitted_curve))
        
    except Exception as e:
        print(f"Could not fit {name} due to error: {e}")

plt.legend()
plt.show()
```

```python
fit_list.sort(key=lambda x: x[1], reverse=True)
for func_name, r_sq, params, fitted_curve in fit_list:
    plt.plot(x_values, y_values, label="data")
    plt.plot(x_values, fitted_curve, label=func_name)
    plt.legend()
    plt.title(f"{func_name}: Params {', '.join(map(str, params))} (R**2 = {r_sq})")
    plt.show()
    print(f"Function: {func_name}, Params: {params} R**2: {r_sq}")
```
